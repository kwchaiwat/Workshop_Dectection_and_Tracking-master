## Linear Regression

```python
df_test = df_pre[['Age','Fare']][:15]
plt.scatter(df_test['Age'],df_test['Fare'])
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = df_test[['Age']]
y = df_test[['Fare']]
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred, color='red',linewidth=2)
```

```python
print('Mean Square Error:'+str(mean_squared_error(y,y_pred)))
print('R2:'+str(r2_score(y,y_pred)))
print('Coefficients:'+str(model.coef_))
```

# การใช้ data มากกว่า 1 ตัว
